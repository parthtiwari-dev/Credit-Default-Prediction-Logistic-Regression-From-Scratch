# EDA

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme(style="whitegrid")
```

```python
df = pd.read_csv("../data/raw.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe().T
```

```python
df["default.payment.next.month"].value_counts()
```

```python
df["default.payment.next.month"].value_counts(normalize=True)
```

```python
sns.countplot(x="default.payment.next.month", data=df)
plt.title("Class Distribution")
plt.savefig("../outputs/figures/class_distribution.png", bbox_inches="tight")
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15,4))

sns.histplot(df["LIMIT_BAL"], kde=True, ax=ax[0])
ax[0].set_title("LIMIT_BAL")

sns.histplot(df["PAY_AMT1"], kde=True, ax=ax[1])
ax[1].set_title("PAY_AMT1")

sns.histplot(df["BILL_AMT1"], kde=True, ax=ax[2])
ax[2].set_title("BILL_AMT1")

plt.tight_layout()
plt.savefig("../outputs/figures/histograms_numeric.png", bbox_inches="tight")
```

```python
pay_cols = [c for c in df.columns if c.startswith("PAY_")]
df[pay_cols].describe().T
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(df.corr(), cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.savefig("../outputs/figures/corr_heatmap.png", bbox_inches="tight")
plt.close()
```

```python
selected = df[["LIMIT_BAL", "AGE", "BILL_AMT1", "PAY_AMT1", "default.payment.next.month"]]
sns.pairplot(selected, hue="default.payment.next.month", diag_kind="kde")
plt.savefig("../outputs/figures/pairplot_selected.png", bbox_inches="tight")
```

```python
edu = df.groupby("EDUCATION")["default.payment.next.month"].mean()
edu
sns.barplot(x=edu.index, y=edu.values)
plt.title("Default Rate by Education")
plt.savefig("../outputs/figures/edu_default.png", bbox_inches="tight")
```

```python
mar = df.groupby("MARRIAGE")["default.payment.next.month"].mean()
sns.barplot(x=mar.index, y=mar.values)
plt.title("Default Rate by Marriage")
plt.savefig("../outputs/figures/marriage_default.png", bbox_inches="tight")
```

```python
bins = [18,30,40,50,60,100]
labels = ['18-29','30-39','40-49','50-59','60+']
df["AGE_BIN"] = pd.cut(df["AGE"], bins=bins, labels=labels, right=False)

ageb = df.groupby("AGE_BIN")["default.payment.next.month"].mean()
sns.barplot(x=ageb.index, y=ageb.values)
plt.title("Default Rate by Age Group")
plt.savefig("../outputs/figures/agebin_default.png", bbox_inches="tight")
```
